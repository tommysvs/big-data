# Grupo Platino

### ¿Qué datos genera?
Describe el tipo, volumen y frecuencia de los datos producidos por el sistema o empresa que has elegido.

> Grupo Platino mediante todas sus empresas genera datos transacciones provenientes de SAP Business One, teniendo un volumen alto de transacciones diarias.

### ¿Quién los usa?
Identifica quiénes son los actores que recopilan, procesan o consumen esos datos: empresas, gobiernos, investigadores, etc.

> El departamento de Finanzas y el departamento de IT de la misma empresa.

### ¿Qué problema resuelve?
Explica cuál es el valor que aporta el análisis de esos datos: ¿mejora un servicio, reduce costes, salva vidas, personaliza experiencias?

> Con estos datos se realiza reporteria por medio de Power Bi, Looker Studio y Crystal Reports, que ayudan a tener información en tiempo real para tomar decisiones y podes ajustar costos o balancear los precios de venta.

### ¿Por qué no bastaría Excel?
Razona en términos de volumen, velocidad, variedad o complejidad por qué una hoja de cálculo o base de datos relacional simple sería 
insuficiente.

> Excel no bastaria, primero porque no es práctico, de primera mano no se recomienda usar Excel como base de datos, y segundo, Grupo Platino usa el motor de base de datos HANA, que es propio de SAP, el cual es súper rapido para obtener datos. De igual manera, se utiliza SQL Server para obtener otros datos del CRM y otros aplicativos hechos en casa.
