```python
# Ejercicio 1: Clasificación de edad

def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 21
categoria = clasificar_edad(edad)
print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")
```

```python
# Ejercicio 3: Calculadora básica

num1 = float(input("Introduce el primer número: "))
num2 = float(input("Introduce el segundo número: "))

print("\nOperaciones permitidas: +, -, *, /")
operacion = input("Introduce el símbolo de la operación: ")

if operacion == "+":
    resultado = num1 + num2
    print(f"\nResultado: {num1} + {num2} = {resultado}")

elif operacion == "-":
    resultado = num1 - num2
    print(f"\nResultado: {num1} - {num2} = {resultado}")

elif operacion == "*":
    resultado = num1 * num2
    print(f"\nResultado: {num1} * {num2} = {resultado}")

elif operacion == "/":
    if num2 == 0:
        print("\nError: No se puede dividir entre cero.")
    else:
        resultado = num1 / num2
        print(f"\nResultado: {num1} / {num2} = {resultado}")

else:
    print("\nError: Operación no válida.")
```

```python
# Ejercicio 4: Número par o impar

numero = int(input("Introduce un número entero: "))

if numero % 2 == 0:
    print(f"El número {numero} es PAR.")
else:
    print(f"El número {numero} es IMPAR.")
```

```python
# Ejercicio 5: Tabla de multiplicar

numero = 7

print(f"Tabla de multiplicar del {numero}:")
for i in range(1, 13):
    print(f"  {numero} x {i} = {numero * i}")
```

```python
# Ejercicio  6: Contador de vocales

frase = "La practica hace al maestroo"
vocales = "aeiou"
contador = 0

for letra in frase.lower():
    if letra in vocales:
        contador += 1

print(f"Frase: {frase}")
print(f"Cantidad de vocales: {contador}")
```

```python
# Ejercicio 7: Promedio de calificaciones

calificaciones = [85, 72, 90, 68, 77]

total = sum(calificaciones)
promedio = total / len(calificaciones)

print(f"Calificaciones: {calificaciones}")
print(f"Suma total: {total}")
print(f"Promedio: {promedio:.1f}")

if promedio >= 70:
    print("Estado: Aprobado")
else:
    print("Estado: Reprobado")
```

```python
# Ejercicio 12: Validación de contraseña

contrasena = "Segura123"

tiene_longitud = len(contrasena) >= 8
tiene_numero = any(c.isdigit() for c in contrasena)
tiene_mayuscula = any(c.isupper() for c in contrasena)
tiene_minuscula = any(c.islower() for c in contrasena)

print(f"Contrasena: {contrasena}")
print(f"  Longitud minima (8): {tiene_longitud}")
print(f"  Contiene numero: {tiene_numero}")
print(f"  Contiene mayuscula: {tiene_mayuscula}")
print(f"  Contiene minuscula: {tiene_minuscula}")

if tiene_longitud and tiene_numero and tiene_mayuscula and tiene_minuscula:
    print("Resultado: Contrasena VALIDA")
else:
    print("Resultado: Contrasena NO valida")
```

```python
# Ejercicio 15: Limpieza básica de texto

frase_original = "   Hola  esto   es una frase de   prueba con   muchos espacios   "
print(f"Frase original: '{frase_original}'\n")

frase_limpia = frase_original.strip()

frase_limpia = frase_limpia.lower()

while "  " in frase_limpia:
    frase_limpia = frase_limpia.replace("  ", " ")

print(f"Frase limpia:   '{frase_limpia}'")

palabras = frase_limpia.split(" ")
total_palabras = len(palabras)
print(f"Número total de palabras: {total_palabras}\n")

print("="*60)
print("EXPLICACIÓN DE LOS MÉTODOS UTILIZADOS:")
print("="*60)
print("- strip(): Elimina todos los caracteres de espacio en blanco (espacios, pestañas, saltos de línea)")
print("           que se encuentren al principio (izquierda) y al final (derecha) de la cadena de texto.")
print("- lower(): Convierte todos los caracteres alfabéticos con mayúsculas de la cadena a minúsculas.")
print("- replace(antiguo, nuevo): Busca una subcadena específica dentro del texto y la reemplaza")
print("                           por otra subcadena nueva todas las veces que aparezca.")
print("- split(separador): Divide una cadena de texto en una lista de subcadenas basándose en un")
print("                    carácter separador especificado (por defecto es cualquier espacio en blanco).")
```

```python
# Ejercicio 16: DataFrame de estudiantes

import pandas as pd

datos_estudiantes = {
    'nombre': ['Carlos', 'Ana', 'Luis', 'Sofia', 'Fernando', 'Maria', 'Jorge', 'Elena', 'Ricardo', 'Beatriz'],
    'carrera': ['Sistemas', 'Medicina', 'Sistemas', 'Derecho', 'Industrial', 'Medicina', 'Sistemas', 'Derecho', 'Industrial', 'Sistemas'],
    'edad': [21, 23, 20, 22, 24, 21, 22, 20, 25, 23],
    'promedio': [85.5, 92.0, 78.0, 88.5, 65.0, 90.0, 81.5, 74.0, 89.0, 95.5],
    'ciudad': ['Tegucigalpa', 'San Pedro Sula', 'Tegucigalpa', 'La Ceiba', 'Choluteca', 'San Pedro Sula', 'Comayagua', 'Tegucigalpa', 'La Ceiba', 'Santa Rosa']
}

df = pd.DataFrame(datos_estudiantes)

print("--- 3. Muestre las primeras 5 filas ---")
display(df.head(5)) 

print("\n--- 4. Muestre la información general del DataFrame ---")
df.info()

print("\n--- 5. Calcule el promedio general de calificaciones ---")
promedio_general = df['promedio'].mean()
print(f"El promedio general de todos los estudiantes es: {promedio_general:.2f}")

print("\n--- 6. Muestre estudiantes con promedio mayor o igual a 80 ---")
estudiantes_altos = df[df['promedio'] >= 80]
display(estudiantes_altos)

print("\n--- 7. Ordene estudiantes por promedio de mayor a menor ---")
df_ordenado = df.sort_values(by='promedio', ascending=False)
display(df_ordenado)
```
