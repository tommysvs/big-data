```python
# Ejercicio 1: Clasificación de edad

def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 21
categoria = clasificar_edad(edad)
print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")
```

```python
# Ejercicio 2: Conversor de temperatura

def convertirt(celsius):
    fahrenheit = (celsius * 9 / 5) + 32
    return fahrenheit

celsius = 15 
resultadof = convertirt(celsius)
print(f"Temperatura: {celsius}")
print(f"Resultado:  {resultadof} °F")
```

```python
# Ejercicio 3: Calculadora básica

num1 = float(input("Introduce el primer número: "))
num2 = float(input("Introduce el segundo número: "))

print("\nOperaciones permitidas: +, -, *, /")
operacion = input("Introduce el símbolo de la operación: ")

if operacion == "+":
    resultado = num1 + num2
    print(f"\nResultado: {num1} + {num2} = {resultado}")

elif operacion == "-":
    resultado = num1 - num2
    print(f"\nResultado: {num1} - {num2} = {resultado}")

elif operacion == "*":
    resultado = num1 * num2
    print(f"\nResultado: {num1} * {num2} = {resultado}")

elif operacion == "/":
    if num2 == 0:
        print("\nError: No se puede dividir entre cero.")
    else:
        resultado = num1 / num2
        print(f"\nResultado: {num1} / {num2} = {resultado}")

else:
    print("\nError: Operación no válida.")
```

```python
# Ejercicio 8: Lista de números positivos y negativos

numeros = [8, -9, 23, -15, 0, 9, 12]
positivos = []
negativos = []

for num in numeros:
    if num > 0:
        positivos.append(num)
    elif num < 0:
        negativos.append(num)

print(f"Todos los números: {numeros}")
print(f"Números positivos: {positivos}")
print(f"Números negativos: {negativos}")

# La lógica de este ejercicio se basa primero en crear la lista de numeros, crear otras dos listas donde se irán guardando los números positivos y negativos, 
# y luego se recorrera la lista de numeros utilizando un ciclo for y evaluando con un if si el numero actual es mayor o menor que cero, para luego añadirlo a 
# la lista que corresponda
```

```python
# Ejercicio 11: Función para calcular descuento

def calcular(precio, porcentaje):
    monto = precio * ( porcentaje /100)
    precio_final = precio - monto 
    return precio_final 
precio_inicial = 700
descuento= 10
pago= calcular (precio_inicial, descuento)
print(f"Precio original: Lps. {precio_inicial}")
print(f"Descuento: {descuento}%")
print(f"Precio final: Lps.{pago}")
```

```python
# Ejercicio 13: Sistema simple de inventario

productos = [
    {"codigo": "INS-0001", "nombre": "Cinta de aislamiento", "cantidad": 100, "precio": 5.00},
    {"codigo": "INS-0002", "nombre": "Cable de red", "cantidad": 50, "precio": 100.00},
    {"codigo": "INS-0003", "nombre": "Conector RJ45", "cantidad": 200, "precio": 2.00}
]

for producto in productos:
    print(f"Código: {producto['codigo']}")
    print(f"Nombre: {producto['nombre']}")
    print(f"Cantidad: {producto['cantidad']}")
    print(f"Precio: L {producto['precio']:.2f}\n")

codigo = "INS-0002"
for producto in productos:
    if producto["codigo"] == codigo:
        print(f"Producto encontrado: {producto['nombre']} - Cantidad: {producto['cantidad']} - Precio: L {producto['precio']:.2f}")
        break

total_inventario = 0
for producto in productos:
    total_inventario += producto["cantidad"] * producto["precio"]

print(f"Valor total del inventario: L {total_inventario:.2f}")

# La lógica del código se basa en recorrer la lista de productos para primero, mostrar su información, luego buscar un producto en base a su 
# código y por último poder realizar el calculo del valor total del inventario.
```

```python
# Ejercicio 14: Registro de ventas

ventas = [
    {"cliente": "Tommy Vega", "producto": "Cinta de aislamiento", "cantidad": 10, "precio_unitario": 5.00},
    {"cliente": "Nelsy Alonzo", "producto": "Cable de red", "cantidad": 5, "precio_unitario": 100.00},
]

total_venta = 0
for venta in ventas:
    total_venta += venta["cantidad"] * venta["precio_unitario"]
    print(f"Cliente: {venta['cliente']} - Total: L {total_venta:.2f}")

total_general = 0
for venta in ventas:
    total_general += venta["cantidad"] * venta["precio_unitario"]

compra_mas_alta = 0
cliente_mas_alto = ""
for venta in ventas:
    total_venta = venta["cantidad"] * venta["precio_unitario"]
    if total_venta > compra_mas_alta:
        compra_mas_alta = total_venta
        cliente_mas_alto = venta["cliente"]
    
print(f"\nTotal general: L {total_general:.2f}")
print(f"Cliente con compra más alta: {cliente_mas_alto} - L {compra_mas_alta:.2f}\n")

# La lógica en este ejercicio se basa en crear una lista con dos ventas y luego recorrer la lista tres veces para sacar tres datos diferentes
# El primero recorre la lista para sacar el total de venta por cliente usando un ciclo for
# El segundo hace lo mismo, solamente que saca el dato de la venta general
# Ya en el tercero es un poco más completo, porque necesito sacar dos datos, la compra mas alta y el cliente de dicha compra
# Para ello, recorro la lista de ventas, voy sumando las ventas de cada uno, y luego evaluo si el total anterior es mayor o menor que la venta actual
# y almaceno el valor de la compra mas alta con su respectivo cliente en dos variables diferentes
```

```python
# Ejercicio 15: Limpieza básica de texto

frase_original = "   Hola  esto   es una frase de   prueba con   muchos espacios   "
print(f"Frase original: '{frase_original}'\n")

frase_limpia = frase_original.strip()

frase_limpia = frase_limpia.lower()

while "  " in frase_limpia:
    frase_limpia = frase_limpia.replace("  ", " ")

print(f"Frase limpia:   '{frase_limpia}'")

palabras = frase_limpia.split(" ")
total_palabras = len(palabras)
print(f"Número total de palabras: {total_palabras}\n")

print("="*60)
print("EXPLICACIÓN DE LOS MÉTODOS UTILIZADOS:")
print("="*60)
print("- strip(): Elimina todos los caracteres de espacio en blanco (espacios, pestañas, saltos de línea)")
print("           que se encuentren al principio (izquierda) y al final (derecha) de la cadena de texto.")
print("- lower(): Convierte todos los caracteres alfabéticos con mayúsculas de la cadena a minúsculas.")
print("- replace(antiguo, nuevo): Busca una subcadena específica dentro del texto y la reemplaza")
print("                           por otra subcadena nueva todas las veces que aparezca.")
print("- split(separador): Divide una cadena de texto en una lista de subcadenas basándose en un")
print("                    carácter separador especificado (por defecto es cualquier espacio en blanco).")
```

```python
# Ejercicio 18: Limpieza de datos con pandas

import pandas as pd

clientes = [
    {"nombre": "Tommy Vega", "edad": 30, "email": None, "ciudad": "San Pedro Sula", "telefono": "98765432"},
    {"nombre": "Nelsy Alonzo", "edad": 24, "email": "nelsy.alonzo@gmail.com", "ciudad": "Puerto Cortés", "telefono": None},
    {"nombre": "Marlon Rivera", "edad": 29, "email": "marlon.rivera@gmail.com", "ciudad": None, "telefono": "92366718"}
]

df_clientes = pd.DataFrame(clientes)

print(df_clientes.isnull().sum())

print("\nValores nulos:", df_clientes[df_clientes.isnull().any(axis=1)].shape[0])

df_clientes["ciudad"] = df_clientes["ciudad"].fillna("No especificada")

df_limpio = df_clientes.dropna(subset=["email"])

display(df_limpio)

# La lógica en este ejercicio es crear primero un diccionario de los clientes y luego pasarlo a un DataFrame, posterior a ello se solicitan varias cosas
# Con la función isnull() se identifican lso valores nulos, y luego solo se cuentan con la funcion de sum()
# Luego para contar los valores nulos por columna se utiliza la misma funcion de isnull() pero con un filtro para contar solo las filas que tengan al menos un valor nulo
# Ya con la funciona fillna() se reemplaza el valor nulo de la columna ciudad por "No especificada"
# Y por último con la función dropna() se eliminan las filas que tengan un valor nulo en la columna email
```

```python
# Ejercicio 19: Análisis de préstamos

import pandas as pd
datos_prestamos = {
    "cliente":         ["Ana Torres", "Luis Mejia", "Maria Paz", "Carlos Vega",
                        "Sofia Ruiz", "Jorge Lima", "Valeria Cruz", "Diego Reyes",
                        "Fernanda Lopez", "Andres Diaz"],
    "monto_solicitado":[45000, 75000, 30000, 120000, 55000,
                        20000, 90000, 48000, 65000, 15000],
    "plazo_meses":     [24, 36, 12, 48, 24, 12, 36, 24, 36, 12],
    "ingreso_mensual": [18000, 25000, 12000, 40000, 20000,
                        9000, 35000, 17000, 22000, 8000],
    "estado":          ["Aprobado", "Aprobado", "Rechazado", "Aprobado",
                        "En revision", "Rechazado", "Aprobado", "En revision",
                        "Aprobado", "Rechazado"]
}

df_pre = pd.DataFrame(datos_prestamos)
print("DataFrame de prestamos:")
print(df_pre)

print(f"\nMonto promedio solicitado: L.{df_pre['monto_solicitado'].mean():,.2f}")

print("\nConteo de solicitudes por estado:")
print(df_pre["estado"].value_counts())

print("\nPrestamos aprobados:")
print(df_pre[df_pre["estado"] == "Aprobado"])

print("\nClientes con monto mayor a L.50,000:")
print(df_pre[df_pre["monto_solicitado"] > 50000])

print("\nPromedio de monto solicitado por estado:")
print(df_pre.groupby("estado")["monto_solicitado"].mean().reset_index())
```

```python
# Ejercicio 20: Mini análisis de datos tipo empresa

import pandas as pd

empleados = [
    {"nombre": "Tommy Vega", "departamento": "IT", "cargo": "Analista de Sistemas", "salario": 33000, "años_experiencia": 7, "ciudad": "San Pedro Sula"},
    {"nombre": "Nelsy Alonzo", "departamento": "Operaciones", "cargo": "Especialista en Digitación", "salario": 18000, "años_experiencia": 2, "ciudad": "Puerto Cortés"},
    {"nombre": "Marlon Rivera", "departamento": "IT", "cargo": "Desarrollador Senior", "salario": 42000, "años_experiencia": 6, "ciudad": "San Pedro Sula"}
 ]

df_empleados = pd.DataFrame(empleados)
display(df_empleados)

print("\nEstadísticas generales de salarios")
estadisticas_salarios = df_empleados["salario"].describe()
display(estadisticas_salarios)

print("\nSalario promedio por departamento")
promedio_por_depto = df_empleados.groupby("departamento")["salario"].mean()
display(promedio_por_depto)

print("\nEmpleado con salario más alto")
top_empleado = df_empleados.loc[df_empleados["salario"].idxmax()]
display(top_empleado)

print("\nEmpleados con más de 5 años de experiencia")
df_mas_experiencia = df_empleados[df_empleados["años_experiencia"] > 5]
display(df_mas_experiencia)

print("\nEmpleados ordenados por salario")
df_ordenado = df_empleados.sort_values(by="salario", ascending=False)
display(df_ordenado)

def nivel_experiencia(años):
    if años < 1:
        return "Junior"
    elif años <= 3:
        return "Intermedio"
    else:
        return "Senior"

df_empleados["NivelExperiencia"] = df_empleados["años_experiencia"].apply(nivel_experiencia)
display(df_empleados)

# La lógica en este ejercicio de igual forma se basa en crear un diccionario primero y luego convertirlo a un DataFrame
# Luego con la funcion de describe() se pueden obtener las estadisticas generales de la columna de salarios
# Para el salario promerio por departamente se utiliza primero la funcion groupby() y luego se obtiene el promedio con la funcion mean()
# Para obtener el empleado con el salario mas alto se utilia la funcion idxmx() que nos permite por medio del id obtener el salario mas alto de X empleado
# y utilizando la funcion loc() se obttiene toda la fila del empleado
# La del empeado con mas de 5 años de experiencia simplemente se evalua la columan de años de experiencia que sea mayor a 5
# Para ordenar los empleados por salario, se utiliza la funcion sort_values() y se ordena de forma descendente
# Y ya por último creamos primero la defincion para saber que se considera Junior, Interemedio o Senior, luego ya con la funcion apply() se aplica esa función 
# a la columna de años de experiencia para crear una nueva columna con el nivel de experiencia de cada empleado
```
